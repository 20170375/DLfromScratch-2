### 2.4 통계 기반 기법 개선하기

#### 2.4.1 상호정보량

$ PMI(x,y) = \log_2 {{P(x,y)} \over {P(x)P(y)}} = \log_2 {{{C(x,y)} \over {N}} \over {{C(x)} \over {N}} {{C(y)} \over {N}}} 
 = \log_2 {{C(x,y) \cdot N} \over {C(x) C(y)}} $
 
 
$ PPMI(x,y) = max(0, PMI(x,y)) $

```python
import numpy as np, sys
sys.path.append('exam_code')
from common.util import preprocess, create_co_matrix, cos_similarity, ppmi

text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus, vocab_size)
W = ppmi(C)

np.set_printoptions(precision=3) # 유효 자릿수를 세 자리로 표시
print('동시발생 행렬\n', C, '\n', '-'*50, '\nPPMI\n', W, '\n')
```

#### 2.4.2 차원감소

**특잇값분해**<sup>Singular Value Docomposition</sup>(SVD)

$ X = USV^T $




#### 2.4.3 SVD에 의한 차원 감소

```python
import numpy as np, sys
sys.path.append('exam_code')
from common.util import preprocess, create_co_matrix, cos_similarity, ppmi

text = 'You say goodbye and I say hello.'
corpus, word_to_id, id_to_word = preprocess(text)
vocab_size = len(word_to_id)
C = create_co_matrix(corpus, vocab_size, window_size=1)
W = ppmi(C)

# SVD는 넘파이의 linalg 모듈에서 제공하는 
# svd 메서드로 실행할 수 있습니다.
U, S, V = np.linalg.svd(W)
```

```python
import matplotlib.pyplot as plt

for word, word_id in word_to_id.items():
    plt.annotate(word, (U[word_id, 0], U[word_id, 1]))
plt.scatter(U[:,0], U[:,1], alpha=0.5)
plt.show()
```

#### 2.4.5 PTB 데이터셋 평가

```python
import numpy as np, sys
sys.path.append('exam_code')
from common.util import most_similar, create_co_matrix, ppmi
from dataset import ptb

window_size  = 2
wordvec_size = 100

corpus, word_to_id, id_to_word = ptb.load_data('train')
vocab_size = len(word_to_id)
print('동시발생 수 계산 ...')
C = create_co_matrix(corpus, vocab_size, window_size=1)
print('PPMI 계산')
W = ppmi(C, verbose=True)

print('SVD 계산 ...')
try:
    # truncated SVD (빠르다!)
    from sklearn.utils.extmath import randomized_svd
    U, S, V = randomized_svd(W, n_components=wordvec_size, n_iter=5, random_state=None)
except ImportError:
    # SVD(느리다)
    U, S, V = np.linalg.svd(W)

word_vecs = U[:, :wordvec_size]

querys = ['you', 'year', 'car', 'toyota']
for query in querys:
    most_similar(query, word_to_id, id_to_word, word_vecs, top=5)
```
